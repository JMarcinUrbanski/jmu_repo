#### Obliczenie prędkości wiatru ze zmiennych u10 & v10 ze wzoru tj z równania pitagorasa
- zmienna u10 => prędkość wiatru zachodniego 
- zmienna v10 => prędkość wiartu północnego 

$$
v_w = \sqrt{u_{10}^2 + v_{10}^2}
$$


```python
# przeliczenie prędkości wiatru 'v_w' dla wszyskich DataFrame pogodowych i dodanie jako nowa kolumna z tymi wartościami

w_ld_2004['v_w'] = np.sqrt(w_ld_2004['u10']**2 + w_ld_2004['v10']**2)
w_ld_2024['v_w'] = np.sqrt(w_ld_2024['u10']**2 + w_ld_2024['v10']**2)
w_pz_2004['v_w'] = np.sqrt(w_pz_2004['u10']**2 + w_pz_2004['v10']**2)
w_pz_2024['v_w'] = np.sqrt(w_pz_2024['u10']**2 + w_pz_2024['v10']**2)
```

```python
# ułożenie kolejnosci kolumn [info data | zmienne pogodowe] 

col_order = ['city','time','year','month','day','day_of_week','u10', 'v10','v_w','t2m','blh','sp','tcc']
w_ld_2004 = w_ld_2004[col_order]
w_ld_2024 = w_ld_2024[col_order]
w_pz_2004 = w_pz_2004[col_order]
w_pz_2024 = w_pz_2024[col_order]
```
