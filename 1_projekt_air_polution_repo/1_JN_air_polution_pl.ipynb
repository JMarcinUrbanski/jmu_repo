## LISTA ZANIECZYSZCEŃ  

| Zanieczyszczenie | Poziom szkodliwości | Dlaczego jest groźne | Główne skutki zdrowotne |
|------------------|--------------------|----------------------|--------------------------|
| PM2.5            | Bardzo wysoki      | Bardzo drobne cząstki – przenikają do krwi | Choroby serca, udary, nowotwory, skrócenie życia |
| Benzo(a)piren    | Bardzo wysoki      | Silnie rakotwórczy składnik smogu | Nowotwory (szczególnie płuc) |
| NO2              | Wysoki             | Drażni drogi oddechowe | Astma, zapalenie oskrzeli, pogorszenie funkcji płuc |
| PM10             | Wysoki             | Wnika do płuc (ale nie do krwi jak PM2.5) | Problemy oddechowe, kaszel, alergie |
| SO2              | Średnio-wysoki     | Silnie drażniący gaz | Skurcze oskrzeli, problemy z oddychaniem |
| O3 (ozon)        | Średnio-wysoki     | Silny utleniacz | Podrażnienia płuc, duszności, spadek wydolności |
| CO               | Wysoki (ostry)     | Blokuje transport tlenu we krwi | Zatrucia, zawroty głowy, śmierć w dużym stężeniu |
| NH3              | Średni             | Drażniący gaz | Podrażnienie oczu i dróg oddechowych |

| Zanieczyszczenie | Główne źródło (działalność człowieka) | Szczegóły |
|------------------|----------------------------------------|----------|
| PM2.5            | Ogrzewanie domów, transport, przemysł  | Spalanie węgla i drewna, spaliny, drobny pył wtórny |
| PM10             | Ogrzewanie domów, transport, przemysł  | Pył z pieców, ścieranie opon i hamulców, kurz |
| NO (tlenek azotu) | Transport, przemysł                   | Spalanie paliw w silnikach i instalacjach przemysłowych |
| NO2 (dwutlenek azotu) | Transport, przemysł              | Utlenianie NO, emisje z silników diesla |
| SO2 (dwutlenek siarki) | Przemysł, energetyka            | Spalanie węgla i paliw zawierających siarkę |
| CO (tlenek węgla) | Transport, ogrzewanie domów           | Niecałkowite spalanie paliw |
| NH3 (amoniak)     | Rolnictwo                             | Nawozy sztuczne, hodowla zwierząt |
| O3 (ozon troposferyczny) | Powstaje wtórnie (transport, przemysł) | Reakcje NOx i LZO pod wpływem słońca |
| Benzo(a)piren     | Ogrzewanie domów                      | Spalanie węgla i drewna, szczególnie w starych piecach |

# 2023 Dane z zanieczyszczeń PL ze strony GIOS.GOV.PL

```python
ds_pm10_2023 = pd.read_excel('/Users/marcinurbanski/Desktop/jmu_repo/1_projekt_air_polution_repo/data_air_pl_2023/2023_PM10_1g.xlsx')
ds_pm25_2023 = pd.read_excel('/Users/marcinurbanski/Desktop/jmu_repo/1_projekt_air_polution_repo/data_air_pl_2023/2023_PM25_1g.xlsx')
ds_no2_2023 = pd.read_excel('/Users/marcinurbanski/Desktop/jmu_repo/1_projekt_air_polution_repo/data_air_pl_2023/2023_NO2_1g.xlsx')
ds_co_2023 = pd.read_excel('/Users/marcinurbanski/Desktop/jmu_repo/1_projekt_air_polution_repo/data_air_pl_2023/2023_CO_1g.xlsx')
```

#### Tworzenie DataFrame dla danych PM10 2023

```python
ds_pm10_2023.columns = ds_pm10_2023.iloc[0]
df_pm10_2023 = ds_pm10_2023.iloc[5:]
df_pm10_2023 = df_pm10_2023[['Kod stacji','LdLodzCzerni','WpPoznSzwajc']]
df_pm10_2023 = df_pm10_2023.rename(columns={'Kod stacji':'Date_time',
                                            'LdLodzCzerni':'LdLodzCzerni_PM10',
                                            'WpPoznSzwajc':'WpPoznSzwajc_PM10'})
df_pm10_2023['Date_time'] = pd.to_datetime(df_pm10_2023['Date_time'])

print(df_pm10_2023.info())
print('\n','='*140,'\n')
display(ds_pm10_2023.head())
display(df_pm10_2023.head(2))
```

#### Tworzenie DataFrame dla danych PM25 2023

```python
ds_pm25_2023.columns = ds_pm25_2023.iloc[0]
df_pm25_2023 = ds_pm25_2023.iloc[5:]
df_pm25_2023 = df_pm25_2023[['Kod stacji','LdLodzCzerni','WpPoznSzwajc']]
df_pm25_2023 = df_pm25_2023.rename(columns={'Kod stacji':'Date_time',
                                           'LdLodzCzerni':'LdLodzCzerni_PM25',
                                            'WpPoznSzwajc':'WpPoznSzwajc_PM25'})
df_pm25_2023['Date_time'] = pd.to_datetime(df_pm25_2023['Date_time'])

print(df_pm25_2023.info())
print('\n','='*140,'\n')
display(ds_pm25_2023.iloc[5:7])
display(df_pm25_2023.head(2))
```

#### Tworzenie DataFrame dla danych NO2 2023

```python
ds_no2_2023.columns = ds_no2_2023.iloc[0]
df_no2_2023 = ds_no2_2023.iloc[6:]
df_no2_2023 = df_no2_2023[['Kod stacji','LdLodzCzerni','WpPoznSzwajc']]
df_no2_2023 = df_no2_2023.rename(columns={'Kod stacji':'Date_time',
                                          'LdLodzCzerni':'LdLodzCzerni_NO2',
                                          'WpPoznSzwajc':'WpPoznSzwajc_NO2'})
df_no2_2023['Date_time'] = pd.to_datetime(df_no2_2023['Date_time'])
print(df_no2_2023.info())
print('\n','='*140,'\n')
display(ds_no2_2023.iloc[6:8])
display(df_no2_2023.head(2))
```

#### Tworzenie DataFrame dla danych CO 2023

```python
# UWAGA: Dla pomiarów CO w Poznaniu w 2024 nie ma danych ze stacji WpPoznSzwajc[2024] tylko sa ze stacji WpPoznDobrow[2023]
#        Dla usprawnienia daslszych agregacji  zapisujemy nazwę stacji w formie uproszczonej WpPozn_CO (bez rozróżnienia
#        która stacja odpowiada za pomiar CO w latach 2023 i 2024)

ds_co_2023.columns = ds_co_2023.iloc[0]
df_co_2023 = ds_co_2023.iloc[5:]
df_co_2023 = df_co_2023[['Kod stacji','LdLodzJanPaw','WpPoznSzwajc']]
df_co_2023 = df_co_2023.rename(columns={'Kod stacji':'Date_time',
                                        'LdLodzJanPaw':'LdLodzJanPaw_CO',
                                        'WpPoznSzwajc':'WpPozn_CO'})
df_co_2023['Date_time'] = pd.to_datetime(df_co_2023['Date_time'])
print(df_co_2023.info())
print('\n','='*140,'\n')
display(ds_co_2023.iloc[5:7])
display(df_co_2023.head(2))
```

#### Merge Wszyskich DataSetów 2023 + Eksport do cvs

```python
from functools import reduce

dfs_list = [df_pm10_2023, df_pm25_2023, df_no2_2023, df_co_2023]

df_polution_2023_merged = reduce(lambda left, right: left.merge(right, on="Date_time", how="left"), dfs_list)

# Eksport do pliku w raw data
df_polution_2023_merged.to_csv('/Users/marcinurbanski/Desktop/jmu_repo/1_projekt_air_polution_repo/0_ready_raw_data/data_polution_2023')
```
