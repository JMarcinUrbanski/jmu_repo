#### Preprocesing

```python
from sklearn.compose import ColumnTransformer
# pozwala stosować różne transformacje do różnych kolumn w zbiorze danych.
# Pozwala:
# wybrać konkretne kolumny,
# przypisać im odpowiednie transformacje,
# połączyć wszystko w jeden przekształcony zbiór danych.
# Dzięki temu możesz:
# ✔ skalować kolumny numeryczne
# ✔ kodować kolumny kategoryczne
# ✔ usuwać wybrane kolumny
# ✔ zostawiać niektóre bez zmian

from sklearn.pipeline import Pipeline
# pozwala połączyć wiele kroków przetwarzania danych oraz model w jeden spójny obiek


from sklearn.preprocessing import OneHotEncoder ,StandardScaler 
# OneHot... -> podział kategorycznych wyników na [0,1] wartosci np male/felamle 0/1.
# Standa... -> do oblicznie z-wartości krytyczne na rozkładzie normalnym 

from sklearn.impute import SimpleImputer
# o klasa ze scikit-learn, służąca do uzupełniania brakujących wartości (NaN) w danych.
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer

num_features = X_train.select_dtypes(include="number").columns
cat_features = X_train.select_dtypes(exclude="number").columns

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler()),
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore")),
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, num_features),
        ("cat", categorical_transformer, cat_features),
    ],
    remainder="drop",
    verbose_feature_names_out=False,
)

print("Num features:", list(num_features))
print("Cat features:", list(cat_features))
```

### Model : Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

# pipeline łaczy wszystkie zapisane kroki w jedna całość, aby uniknac pomylek 
# najpier wykona się preproccesing a nastepnie model zostanie wytrenowany.
clf = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('logreg', LogisticRegression(max_iter=2000))
    # domyśłnie jest 100 iteracji 
])

# tronowanie modelu: imputacja braków / skalowanie / one-hot encoding -> powstaje macierzch numeryczna
clf.fit(X_train, y_train)

# predykcja na train
y_pred = clf.predict(X_test)
y_proba = clf.predict_proba(X_test)[:,1]

print('Model Wytrenowany')
```

### Metryki (próg 0.5)

#### Confussion_matrix
to tabela gdzie:
- TP – przeżył i model przewidział przeżycie
- TN – nie przeżył i model przewidział brak przeżycia
- FP – model powiedział „przeżył”, ale nie przeżył
- FN – model powiedział „nie przeżył”, ale przeżył

#### accuracy_score
$Accuracy =  \frac{TP+TN}{TP+TN+FP+FN}$

Procent poprawnych predykc

#### recall(czułość) 
odpowiada na pytania ile faktycznie ocalałych model wykrył?

$ Recall = \frac{TP}{TP+FN}$

#### f1 
To średnia harmoniczna preccision i recall

$F1 = 2*\frac{Precision*Recall}{Precision+Recall}$


#### classification_report

```python
from sklearn.metrics import (confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score)

cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix: \n", cm)

acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, zero_division=0)
rec = recall_score(y_test, y_pred, zero_division=0)
f1 =  f1_score(y_test, y_pred, zero_division=0)

print(f'Accuracu: {acc:.3f}') # ilość poprawnych predykcji 
print(f'Precision: {prec:.3f}') # Gdy model mówi 'przezył' -> ma rację w ilu %
print(f'Recall: {rec:.3f}') # Wykrywa ile % faktycznie ocalałych 
print(f'F1: {f1:.3f}') # średnia harmoniczna z prec i receall ->
print("\nclassification_report:")
print(classification_report(y_test, y_pred, zero_division=0))

```

### Price_Change: znak i wartość

```python
s = df['Price_Change'].astype(str)

# na poczatku toworzymy na podstwaie kolumny 'price_chanfe' mnożnik 1 lub -1 . 
# gdy zawiera 'trading_down' to wartosć jest -1 a w przeciwnym razie +1. po przez metode .map()
# nastepnie wyciafamy z calego wyrazenia same cyfry (liczbe) za pomowca regex
# mnozenie pozwala zapisac zmiane wartosci ceny z wartoscia '-' jeżeli był spadek lub '+' jezeli byl wzrost , calosc convert na float

df['Price_Change_Value'] = (
    s.str.contains('trading_down', na=False)
    .map({True: -1 , False: 1})
    * 
    s.str.replace(r'[^0-9]','',regex=True)
    .replace('',pd.NA)
    .astype(float)
)

df[df['Price_Change_Value'] < 0]

display(df[["Price_Change","Price_Change_Value"]] )
```

### Braki danych + imputacja 

#### Outliery

```python
candidate_cols = [c for c in ["New_Price", "Old_Price", "Price_Change_Value"] if c in df.columns]

# funkcja sprwdzajca sume outlierów dla wybranych kolumn, wykraczjacych poza minimum i maximum z 1,5*IQR 
def count_outliers_iqr(s):
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1 
    lo, hi = q1 - 1.5*iqr, q3 +1.5*iqr
    return ((s<lo) | (s>hi)).sum()


for c in candidate_cols:
    n = count_outliers_iqr(df[c].dropna()) # .dropna() -> usuwa brakujce dane 
    print(f'{c}: {n} outliers')

```

```python
import matplotlib.pyplot as plt

# wykresy boxplot dokłdnie pokazuja skale oulierów, czyli czarnych kółek poza zakresem IQR[+-]*1,5

plt.figure(figsize=(8,4))
plt.boxplot(df[candidate_cols])
plt.title("Boxplot - detekcja outlierów")
plt.grid()
plt.show()
```

## Opcjonalnie winsoryzacja

```python
# tworzymy kopie dla df, aby nie nadpisywać oryginalnego DataFrame 
df_wins = df.copy()

# petla for dla wszyskich kolumn numerycznych (ceny) stosuje przypisanie granicznej wartosci 1% lub 99 % obserwacji, 
# dla niższych i wyższych wartosći 

# .quantile(0.01)-> wartość poniżej której lezy 1% obserwacji 
# .quantile(0.99)-> wartość poniżej której lezy 99% obserwacji 

# metoda .clip() przypisuje niższe i wyżesze wartosći do wskazancyh zmienych. i zamienia je na te wartości, 
# w tym przypadku na zmienne lo i hi.

for c in candidate_cols:
    lo,hi = df_wins[c].quantile(0.01), df_wins[c].quantile(0.99)
    df_wins[c] = df_wins[c].clip(lower=lo, upper=hi)

print("Po winsoryzacji:")
display(df_wins[candidate_cols].describe().T)
```
